# Differences between Lasso and Ridge

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Data for the table
data = {
    "": ["Regularization / Penalty term", "Constraint surface", "Coefficients after Regularization", "Feature Selection Method"],
    "Lasso Regularization": [
        "L1 norm (sum of absolute values) of coefficients",
        "Diamond shaped in 2D",
        "All the coefficients shrink in their magnitude and some can be zero.",
        "Yes! Features corresponding to the zero value coefficients are eliminated."
    ],
    "Ridge Regularization": [
        "L2 norm (sum of squared values) of coefficients",
        "Circle shaped in 2D",
        "All the coefficients shrink in their magnitude but never approach zero.",
        "No! But it helps in emphasizing the relative feature importance."
    ]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the table
fig, ax = plt.subplots(figsize=(17, 6))  # Set the size of the table
ax.axis('off')  # Remove the axes

# Render the table
tab = table(ax, df, loc='center', cellLoc='center', colWidths=[0.4]*len(df.columns))
tab.auto_set_font_size(False)
tab.set_fontsize(10)

# Adjust the row height by setting cell properties
for key, cell in tab.get_celld().items():
    cell.set_height(0.1)  # Increase the row height (adjust value as needed)

plt.show()
```

# Hyperparameter Tuning and the Bias-Variance Tradeoff

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Data for the table
data = {
    "": ["Low alpha","High s", "Low Bias","High Variance", "Low Regularization", "Overfitting", "Low training error"],
    " ": ["High alpha","Low s", "High Bias","Low Variance", "High Regularization", "Underfitting", "High training error"]
}

# Create the DataFrame
df = pd.DataFrame(data, index=[" ", " ", " ", " ", " ", " ", " " ])

# Create a matplotlib figure with larger size
fig, ax = plt.subplots(figsize=(12, 6))  # Increase figure size
ax.axis('off')  # Remove the axes

# Render the table with adjustments for size and row height
tab = table(ax, df, loc='center', cellLoc='center', colWidths=[0.4]*len(df.columns))  # Increase column widths
tab.auto_set_font_size(False)
tab.set_fontsize(12)  # Increase font size

# Adjust the row height by setting cell properties
for key, cell in tab.get_celld().items():
    cell.set_height(0.1)  # Set the row height to 0.2 (you can adjust this)

# Display the table
plt.show()
```
